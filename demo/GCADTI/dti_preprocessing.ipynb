```python
import pandas as pd
```

```python
df = pd.read_csv('pred_dti_procan.csv')
```

```python
df
```

```python
df_sorted = df.sort_values(by=['ID', 'sequence'], ascending=[True, True])
```

```python
df_sorted
```

```python
# 그룹별로 sequence 리스트 만들기
group_seq = df_sorted.groupby("ID")['sequence'].apply(list)

# 첫 번째 그룹의 sequence를 기준으로 비교
first_seq = group_seq.iloc[0]
all_equal = group_seq.apply(lambda x: x == first_seq).all()

print("모든 그룹의 sequence 리스트가 동일한가?", all_equal)
```

```python
group_label_by_cid = df_sorted.groupby("ID")['label'].apply(list)
```

```python
import torch
group_label_tensor_dict = {cid: torch.tensor(labels) for cid, labels in group_label_by_cid.items()}
print(len(group_label_tensor_dict))
```
